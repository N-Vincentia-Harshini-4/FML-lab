```python
import numpy as np
import matplotlib.pyplot as plt
def sig(x):
  return 1/(1+np.exp(-x))
x=np.linspace(-10,10)
plt.plot(x,sig(x))
plt.axis("tight")
*plt.title('Activation Function:sigmoid')
plt.show()
```

```python
def tanh(x):
  return np.tanh(x)
x=np.linspace(-10,10)
plt.plot(x,sig(x))
plt.axis('tight')
plt.title('Activation Function:tanh')
plt.show()
```
