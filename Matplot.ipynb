```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
```

```python
y = np.array([25, 45, 15, 10])
plt.pie(y)
plt.show()
```

```python
y = np.array([55, 10, 15, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y, labels = mylabels)
plt.show()
```

```python
y = np.array([65, 15, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```
